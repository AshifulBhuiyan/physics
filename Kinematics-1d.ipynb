## Problem 1
A car is moving along a straight road with a constant speed of $25 \, \text{m/s}$. At $t = 0$, the car is at $x = 0$.

a) Write a program that calculates the position of the car at $t = 3 \, \text{s}$.

b) Write a program that calculates the position of the car at $t = 3 \, \text{s}$, but this time, the car is initially at $x = 10 \, \text{m}$.

c) Write a program that calculates the position of the car at $t = 3 \, \text{s}$, but this time, the car is initially at $x = 10 \, \text{m}$ and has a constant acceleration of $2 \, \text{m/s}^2$.

d) Write a program that calculates the position of the car at $t = 3 \, \text{s}$, but this time, the car is initially at $x = 10 \, \text{m}$ and has a constant acceleration of $2 \, \text{m/s}^2$. The program should also calculate the velocity of the car at $t = 3 \, \text{s}$.

e) Write a program that calculates the position of the car at $t = 3 \, \text{s}$, but this time, the car is initially at $x = 10 \, \text{m}$ and has a constant acceleration of $2 \, \text{m/s}^2$. The program should also calculate the velocity of the car at $t = 3 \, \text{s}$ and the displacement of the car from its initial position.

f) Write a program that calculates the position of the car at $t = 3 \, \text{s}$, but this time, the car is initially at $x = 10 \, \text{m}$ and has a constant acceleration of $2 \, \text{m/s}^2$. The program should also calculate the velocity of the car at $t = 3 \, \text{s}$, the displacement of the car from its initial position, and the average velocity of the car during the time interval from $t = 0$ to $t = 3 \, \text{s}$.


```python
x = 0
v = 0.45
t = 0
dt = 0.25

while t < 1.5:
    print("t = ", t )
    t = t + dt
    x = x + v * dt
    print("x = ", x)

# Prepare data for plotting
times = [0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5]  # Time points from the loop
positions = []  # List to store positions for plotting

# Calculate positions for each time point
for time in times:
    position = 0.45 * time
    positions.append(position)  # Add position to the list  


# Create the plot
plt.plot(times, positions)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs Time')
plt.grid()
plt.show()
```
