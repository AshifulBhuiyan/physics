# Magnetic Field due to a Moving Point Charge

$$
\vec{B} = \frac{\mu_0}{4\pi} \frac{q \vec{v} \times \hat{r}}{|r|^2}
$$

```python
class Charge:
    def __init__(self, x, y, z, q):
        self.r = [x,y,z]
        self.pos = np.array(self.r)
        self.q = q
        self.v = [0,0,0]
```

```python
km=1e-7  # constant for magnetic field calculation
q1 = Charge(0, 0, 0, 6e-6)
q1.v = [0, 0, 222]  # velocity vector of the charge

# Initialize plot
plt.figure(figsize=(8, 8))
theta = 0
N = 8
dtheta = (2 * np.pi) / N  # Fixed delta theta to complete full rotation

R = 0.1
while theta < 2 * np.pi:
    r = np.array([R * np.cos(theta), R * np.sin(theta), 0])
    B = km * q1.q * np.cross(q1.v, r) / np.linalg.norm(r) ** 3
    plt.quiver(r[0], r[1], B[0], B[1], scale=1e-8, color='r')  # Adjust scale for visibility
    theta += dtheta

# Setting plot limits and labels
plt.xlim(-0.15, 0.15)
plt.ylim(-0.15, 0.15)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Magnetic Field Visualization')
plt.axis('equal')
plt.grid()
plt.show()

# Save the plot
```

# Magnetic Field due to a Long Straight Wire

```python
def Bn(rr):
    km = 1e-7
    L = 1
    I = 0.1
    xo = np.linalg.norm(rr)

    N = 400
    dL = np.array([0, L/N, 0])

    rstart = np.array([0,-L/2,0]) + dL/2
    ro = np.array([xo,0,0])
    B = np.zeros(3)


    n=0

    while n<N:
        rp = rstart + n*dL
        #long straight wire
        r = rp - ro
        dB = km*I*np.cross(dL,r)/np.linalg.norm(r)**3
        B += dB
        n += 1
    
    return B


km = 1e-7
L = 1
I = 0.1
rr = np.array([0.05,0,0])
xo = np.linalg.norm(rr)

print("B = ", Bn(rr))
Ba = km*2*I/xo
Bc = km*I*L/(xo*(np.sqrt(xo**2+(L/2)**2)))
print("B from formula = ", Ba)
print("B from calculation = ", Bc)
```

```python
#plot magnitude of B with respect to distance from the wire
r_mag = np.linspace(0.01, 2, 100)
B_mag = np.zeros_like(r_mag)

for i in range(len(r_mag)):
    rr = np.array([r_mag[i],0,0])
    B_mag[i] = np.linalg.norm(Bn(rr))

plt.figure(figsize=(8, 6))
plt.plot(r_mag, B_mag, label='Calculated B')
plt.plot(r_mag, km*2*I/r_mag, label='Analytical B')
plt.xlabel('Distance from Wire (m)')
plt.ylabel('Magnetic Field Magnitude (T)')

plt.title('Magnetic Field of a Long Straight Wire')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
km = 1.0  # Magnetic constant
I = 1.0   # Current in Amperes
R = 0.05  # Radius of the circular wire in meters
N = 100   # Number of segments for numerical calculation
dtheta = 2 * np.pi / N  # Angular segment size

# Function to compute magnetic field using Biot-Savart law

def B(rto):
    B_total = np.zeros(3)
    for i in range(N):
        r = np.array([0, R * np.cos(i * dtheta), R * np.sin(i * dtheta)])
        rnorm = r/np.linalg.norm(r)
        dl = (2 * np.pi * R / N) * np.cross([1, 0, 0], rnorm)
        ro = rto - r
        B_total += km * I * np.cross(dl, ro) / np.linalg.norm(ro)**3
    return B_total

# Range of distances from the wire
r_mag = np.linspace(0.03, 0.1, 100)
B_mag = np.zeros_like(r_mag)   # Analytical magnetic field
B_mag2 = np.zeros_like(r_mag)  # Numerical magnetic field
B_mag3 = np.zeros_like(r_mag)  # Approximate magnetic field

# Compute magnetic fields
for i, r in enumerate(r_mag):
    rr = np.array([r, 0, 0])  # Observation point
    Bn = B(rr)                # Numerical magnetic field
    B_mag[i] = km * 2 * I * np.pi * R**2 / (r**2 + R**2)**(3/2)  # Analytical
    B_mag2[i] = np.linalg.norm(Bn)  # Numerical magnitude
    B_mag3[i] = km * 2 * I * np.pi * R**2 / (r**3)  # Approximation (far field)

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(r_mag, B_mag, label='Analytical B', linestyle='-', linewidth=2)
plt.plot(r_mag, B_mag2, label='Numerical B', linestyle='--', linewidth=2)
plt.plot(r_mag, B_mag3, label='Approximate B', linestyle=':', linewidth=2)
plt.xlabel('Distance from Wire (m)')
plt.ylabel('Magnetic Field Magnitude (T)')
plt.title('Magnetic Field of a Circular Wire')
plt.grid()
plt.legend()
plt.show()
```
