```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Monte Carlo Calculation of Pi

```python
plt.figure(figsize=(10,10))
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.grid(True, which='both')
plt.scatter(0,0,color='black')
plt.scatter(1,0,color='black')
plt.scatter(0,1,color='black')
plt.scatter(-1,0,color='black')
plt.scatter(0,-1,color='black')

N = 10000
n = 0
cin = 0

# numpy make vector

while n < N:
    r = np.random.rand(2) * 2 - 1  # Change to generate points in the entire square
    x = r[0]
    y = r[1]
    if x**2 + y**2 < 1:
        plt.scatter(x,y,color='red')
        cin += 1
    else:
        plt.scatter(x,y,color='green')
    n += 1

tpi = cin * 4 / N  # Note: Adjusted the denominator to N instead of n
print("pi", tpi)
```
